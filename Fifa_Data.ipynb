```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# Read Data From CSV File.
fifa = pd.read_csv('fifa data.csv')
```

```python
# Counts of Players with respective Country.
fifa.nationality.value_counts()
```

```python
# Get Top 5 countries where most football payers belongs
fifa.nationality.value_counts()[0:5].keys()
```

```python
# Bar Plot for Top 5 Countries which has most players
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color='r')
plt.title("Top 5 Country");
plt.xlabel("Country")
plt.ylabel("No of Players")
plt.legend();
```

```python
# Create Datframe with two columns short_name and wage_eur(salary) using original dataframe.
player_salary = fifa[['short_name','wage_eur']]
```

```python
# Sort players by salary in descending ordder.
player_salary = player_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
# print datafrme.
player_salary.head()
```

```python
# Bar Plot for Top 5 Players which has most salary.
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color='orange')
plt.title("Top 5 Highes Paying Salary")
plt.xlabel("Name of Player")
plt.ylabel("Salary")
plt.legend();
```
